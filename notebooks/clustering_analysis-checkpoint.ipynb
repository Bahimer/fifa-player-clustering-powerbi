```python
import pandas as pd

# Charger seulement les 50 000 premières lignes
fichier = r"C:\Users\ibmer\Desktop\archive\male_players.csv"
df = pd.read_csv(fichier, nrows=50000)

# Vérification
print("Dimensions de l'échantillon :", df.shape)
print(df.head())
```

```python
df.to_csv(r"C:\Users\ibmer\Desktop\archive\male_players_sample.csv", index=False)
```

```python
print(df.columns.tolist())
```

```python
columns_to_keep = [
    'short_name',        # Nom du joueur
    'age',
    'nationality_name',
    'club_name',
    'player_positions',
    'overall',
    'potential',
    'pace',
    'shooting',
    'passing',
    'dribbling',
    'defending',
    'physic'
]
```

```python
# On garde uniquement les colonnes choisies
df_filtered = df[columns_to_keep].copy()

# Suppression des lignes incomplètes
df_filtered.dropna(inplace=True)

# Vérification
print("Dimensions après filtrage :", df_filtered.shape)
print(df_filtered.head())
```

```python
import matplotlib.pyplot as plt

# Trier les joueurs par score 'overall' décroissant, puis prendre les 10 premiers
top_overall = df_filtered.sort_values(by="overall", ascending=False).head(10)

# Définir la taille de la figure
plt.figure(figsize=(10, 6))

# Créer un graphique à barres horizontales avec les noms et les scores
plt.barh(top_overall['short_name'], top_overall['overall'], edgecolor='black')

# Inverser l’axe Y pour que le meilleur joueur soit en haut
plt.gca().invert_yaxis()

# Ajouter un titre et un label à l’axe X
plt.title("Top 10 joueurs - Score global (Overall)")
plt.xlabel("Note overall")

# Ajuster automatiquement les marges
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
# Créer une nouvelle figure avec une taille adaptée
plt.figure(figsize=(8, 5))

# Créer un histogramme du score 'shooting' avec 30 intervalles (bins)
plt.hist(df_filtered['shooting'], bins=30, edgecolor='black')

# Ajouter un titre et des étiquettes aux axes
plt.title("Distribution du score Shooting")
plt.xlabel("Shooting")
plt.ylabel("Nombre de joueurs")

# Ajouter une grille au fond du graphique
plt.grid(True)

# Ajuster les marges pour que tout soit visible
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
# Grouper les joueurs par nationalité et calculer la moyenne des stats pour chaque pays
df_nationality = df_filtered.groupby('nationality_name')[['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']].mean()

# Trier les nationalités selon la moyenne en 'shooting' et prendre les 10 meilleures
top_nat = df_nationality.sort_values(by='shooting', ascending=False).head(10)

# Créer un graphique à barres groupées avec les moyennes des différentes stats
top_nat.plot(kind='bar', figsize=(12, 6))

# Ajouter un titre et une étiquette verticale
plt.title("Moyenne des stats des 10 meilleures nationalités (par shooting)")
plt.ylabel("Note moyenne")

# Rotation des noms de pays pour une meilleure lisibilité
plt.xticks(rotation=45)

# Ajuster les marges
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
# On garde uniquement les stats de jeu (features pour le clustering)
X = df_filtered[['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']].copy()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardisation (moyenne = 0, écart-type = 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

# KMeans avec 4 groupes de joueurs (tu peux tester d'autres nombres ensuite)
kmeans = KMeans(n_clusters=4, random_state=42)

# Appliquer le modèle
clusters = kmeans.fit_predict(X_scaled)

# Ajouter le numéro de cluster au dataframe d'origine
df_filtered['cluster'] = clusters
```

```python
# Compter combien de joueurs dans chaque cluster
print(df_filtered['cluster'].value_counts().sort_index())
```

```python
# Moyenne des stats par groupe
cluster_profiles = df_filtered.groupby('cluster')[['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']].mean()
print(cluster_profiles)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Réduction de dimension avec PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Ajout des composantes PCA au DataFrame pour visualisation
df_filtered['pca_1'] = X_pca[:, 0]
df_filtered['pca_2'] = X_pca[:, 1]

# Affichage du scatter plot des clusters
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_filtered['pca_1'], df_filtered['pca_2'],
                      c=df_filtered['cluster'], cmap='tab10', alpha=0.6)

plt.title("Visualisation des clusters de joueurs (PCA)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```
